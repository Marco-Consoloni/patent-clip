# Memory

Memory:
* 1 tar archive = 4GB
* 1 tar of zip file = 4GB
* 1 tar of XML file = 863M
* 1 tar of front img file = 220M
* 1 json for a tar = 40M

Memory Estimates (year):
* 1 full year of tar =  185GB
* 1 full year of zip file = 4GB * 52 = 208GB
* 1 full year of XML file = 863M * 52 = 45GB
* 1 full year of front img file = 220M * 52 = 12GB
* 1 full year of json file = 40M * 52 = 2GB

Memory Estimates (total):
* tar folder = 185GB * 5 = 1TB
* zip folder = 208GB * 5 = 1TB
* XML folder = 45GB * 5 = 225GB
* front img folder = 12GB * 5 = 60GB
* json file = 2GB * 5 = 10GB

Tot = 2.5 TB (memory on disk comprising tar files, zip files)

# Time Analytics

* time to download 1 tar archive = 18-20 min (on average)
* time to download all tar archives from 1 full year = 18-20 min * 52 = 1040 min  
* time to extract 1 zip from 1 tar archive = 50 sec (on average)
* time to extract all zip files from 1 full year = 50 sec * 52 = 45 min  
* time to extract XML and front image of 1 tar = 95 sec
* time to extract XML and front image from 1 full year = 90 sec * 52 = 78 min  
* time to create a JSON file for a tar = 40 sec
* time to create JSON files for a year = 40 sec * 52 =  30 min  
* time to merge all JSON file into one = 180 sec = 3 min
